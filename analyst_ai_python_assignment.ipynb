```python
import requests
from bs4 import BeautifulSoup
import csv

# Initialize the CSV file
csv_file = open('amazon_products.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Product URL', 'Product Name', 'Product Price', 'Rating', 'Number of Reviews'])

# Loop through multiple pages
for page_num in range(1,1000):
    url = f'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{page_num}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract product details from the page
    product_containers = soup.find_all('div', {'data-asin': True})

    for container in product_containers:
        product_link = container.find('a', {'class': 'a-link-normal'})
        product_url = 'https://www.amazon.in' + product_link['href'] if product_link else 'N/A'

        product_name_container = container.find('span', {'class': 'a-size-medium'})
        product_name = product_name_container.text.strip() if product_name_container else 'N/A'

        product_price_container = container.find('span', {'class': 'a-offscreen'})
        product_price = product_price_container.text.strip() if product_price_container else 'N/A'

        rating_container = container.find('span', {'class': 'a-icon-alt'})
        rating = rating_container.text.split()[0] if rating_container else 'N/A'

        num_reviews_container = container.find('span', {'class': 'a-size-base'})
        num_reviews = num_reviews_container.text.strip().replace(',', '') if num_reviews_container else '0'

        csv_writer.writerow([product_url, product_name, product_price, rating, num_reviews])

# Close the CSV file
csv_file.close()
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Load product URLs from the CSV generated in Part 1
product_urls = []
with open('amazon_products.csv', 'r', encoding='utf-8') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)  # Skip the header row
    for row in csv_reader:
        product_urls.append(row[0])  # Assuming product URLs are in the first column

# Initialize the CSV file for product details
csv_file = open('product_details.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Product URL', 'Description', 'ASIN', 'Product Description', 'Manufacturer'])

# Loop through product URLs and fetch details
for product_url in product_urls:
    if product_url == 'N/A':
        csv_writer.writerow(['N/A', 'N/A', 'N/A', 'N/A', 'N/A'])
        continue  # Skip to the next iteration

    response = requests.get(product_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Description
        description = soup.find('meta', {'name': 'description'})
        description = description['content'] if description else 'N/A'

        # ASIN
        asin = soup.find('input', {'id': 'ASIN'})
        asin = asin['value'] if asin else 'N/A'

        # Product Description
        product_description_container = soup.find('div', {'id': 'productDescription'})
        product_description = product_description_container.text.strip() if product_description_container else 'N/A'

        # Manufacturer
        manufacturer_container = soup.find('a', {'id': 'bylineInfo'})
        manufacturer = manufacturer_container.text.strip() if manufacturer_container else 'N/A'

        csv_writer.writerow([product_url, description, asin, product_description, manufacturer])
    else:
        print(f"Failed to fetch details for {product_url}. Status code: {response.status_code}")

# Close the CSV file
csv_file.close()
```
